```python
import numpy as np  
import matplotlib.pyplot as mtp  
import pandas as pd 
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split  
from sklearn.linear_model import LinearRegression  
from sklearn.metrics import mean_squared_error, r2_score

# Load the bike dataset'
bike_details = pd.read_csv("BIKE DETAILS.csv")

# Prepare the data: km_driven as the feature (X) and selling_price as the target (y)
x = bike_details[['km_driven']].dropna()  # Feature (independent variable)
y = bike_details['selling_price'].dropna()  # Target (dependent variable)

# Ensure that the corresponding values in x and y are aligned after dropping NAs
x = x.loc[y.index]

# Split the dataset into the training set and the testing set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1/3, random_state=0)

# Fit the Linear Regression model to the training set
regressor = LinearRegression()
regressor.fit(x_train, y_train)

# Predicting the Test set results
y_pred = regressor.predict(x_test)
x_pred = regressor.predict(x_train)

# Plot the training set results
mtp.scatter(x_train, y_train, color="green")
mtp.plot(x_train, x_pred, color="red")
mtp.title("Linear Regression on Bike Dataset (Training Set)")
mtp.xlabel("Kilometers Driven")
mtp.ylabel("Selling Price")
mtp.show()

# Plot the test set results
mtp.scatter(x_test, y_test, color="blue")
mtp.plot(x_train, x_pred, color="red")
mtp.title("Linear Regression on Bike Dataset (Test Set)")
mtp.xlabel("Kilometers Driven")
mtp.ylabel("Selling Price")
mtp.show()

# Print the coefficients and intercept of the model
print('Coefficients:', regressor.coef_)
print('Intercept:', regressor.intercept_)

# Calculate and print the Mean Squared Error and R^2 Score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print('Mean Squared Error:', mse)
print('Coefficient of Determination (R^2 Score):', r2)
```
